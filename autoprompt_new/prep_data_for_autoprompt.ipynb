```python
IN_FILE = "/home/rbhaskar/nlp/summer_2022/NLP_Project/pararel/correctly_classified_roberta_large.jsonl"
OUT_FILE = "/home/rbhaskar/nlp/summer_2022/NLP_Project/pararel/correctly_classified_roberta_large_autoprompt_format.jsonl"
SHORT_FILE = "/home/rbhaskar/nlp/summer_2022/NLP_Project/pararel/correctly_classified_roberta_large_autoprompt_format_short.jsonl"
SHORTER_FILE = "/home/rbhaskar/nlp/summer_2022/NLP_Project/pararel/correctly_classified_roberta_large_autoprompt_format_shorter.jsonl"
FIVEK_FILE = "/home/rbhaskar/nlp/summer_2022/NLP_Project/pararel/correctly_classified_roberta_large_autoprompt_format_5k.jsonl"
MASK_TOKEN = "<mask>"
SINGLE_ENTITY_FILE = "/home/rbhaskar/nlp/summer_2022/NLP_Project/pararel/correctly_classified_roberta_large_autoprompt_format_single_entity.jsonl"
```

```python
with open(IN_FILE, "r", encoding="utf-8") as f:
    lines = f.readlines()

newlines = []

for line in lines:
    line = json.loads(line)
    text = line.pop("Text")
    pre_mask, post_mask = text.split(MASK_TOKEN)
    line["Pre_Mask"] = pre_mask
    line["Post_Mask"] = post_mask
    newlines.append(json.dumps(line) + "\n")

with open(OUT_FILE, "w", encoding="utf-8") as f:
    f.writelines(newlines)
```

```python
random.seed(0)

with open(OUT_FILE, "r", encoding="utf-8") as f:
    lines = f.readlines()

with open(FIVEK_FILE, "w", encoding="utf-8") as f:
    sample_5k = random.sample(lines, 5000)
    f.writelines((line + "\n" for line in sample_5k))
```

```python
with open(OUT_FILE, "r", encoding="utf-8") as f:
    lines = f.readlines()

from collections import defaultdict

single_entity = defaultdict(lambda: defaultdict(list))

for line in lines:
    row = json.loads(line)
    dataset = row["Dataset"]
    prompt = row["Pre_Mask"] + "[Y]" + row["Post_Mask"]
    template_before_X, template_after_X = row["Prompt"].split("[X]")
    subject = prompt
    if template_before_X not in {"", " "}:
        subject = subject.replace(template_before_X, "")
    if template_after_X not in {"", " "}:
        subject = subject.replace(template_after_X, "")

    if subject == "":
        raise Exception("you messed up something")

    single_entity[dataset][subject].append(line)

```
